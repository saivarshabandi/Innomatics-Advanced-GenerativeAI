```python
def get_unique_words(text):
    text = text.lower()
    for char in '.,!?;:':
        text = text.replace(char, '')
        words = text.split()
        unique_set = set(words)

    return len(unique_set), unique_set
sentence = input("Enter your sentence: ")
if sentence.strip():
    count, unique_words = get_unique_words(sentence)
    print(f"Unique words count: {count}")
    print(f"Unique words: {unique_words}")
else:
    print("The input was empty. Try again!")
```

```python
def find_highest_paid(employees):
    if not employees:
        return None, 0
    highest_name = ""
    highest_salary = 0
    for name, salary in employees.items():

        if salary > highest_salary:
            highest_salary = salary
            highest_name = name

    return highest_name, highest_salary
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}

name, salary = find_highest_paid(employees)

if name:
    print(f"Highest Salary: {name} - {salary}")
else:
    print("No employee data found.")
```

```python
def find_max_min(numbers):
    if not numbers:
        return None, None

    max_val = numbers[0]
    min_val = numbers[0]

    for num in numbers:
        if num > max_val:
            max_val = num
        if num < min_val:
            min_val = num

    return max_val, min_val

if __name__ == "__main__":
    numbers = [45, 22, 89, 10, 66]

    max_val, min_val = find_max_min(numbers)

    print(f"List: {numbers}")
    print(f"Max: {max_val}")
    print(f"Min: {min_val}")
```

```python
def count_expensive_products(prices):
    count = 0

    for price in prices:
        if price > 1000:
            count += 1

    return count

if __name__ == "__main__":
    prices = [450, 1200, 899, 1500, 300]

    total = count_expensive_products(prices)

    print(f"Products above 1000: {total}")
```

```python
def calculate_attendance(attendance_list):
    if not attendance_list:
        return 0.0

    total_days = len(attendance_list)
    present_days = 0

    for status in attendance_list:
        if status == "P":
            present_days += 1

    percentage = (present_days / total_days) * 100
    return percentage

if __name__ == "__main__":
    attendance = ["P", "P", "A", "P", "P"]

    result = calculate_attendance(attendance)

    print(f"Attendance Percentage: {result}")
```

```python
def check_key_exists(data_dict, key_to_find):

    if key_to_find in data_dict:
        return True
    else:
        return False

if __name__ == "__main__":

    employees = {
        "Ravi": 75000,
        "Anita": 68000,
        "Kiran": 72000
    }


    if check_key_exists(employees, "Ravi"):
        print("Employee exists")
    else:
        print("Employee not found")
```

```python
def calculate_average(marks_list):
    if not marks_list:
        return 0.0

    total_sum = sum(marks_list)
    count = len(marks_list)

    average = total_sum / count
    return average

if __name__ == "__main__":
    marks = [85, 70, 60, 75]

    result = calculate_average(marks)

    print(f"Average Marks: {result}")
```
