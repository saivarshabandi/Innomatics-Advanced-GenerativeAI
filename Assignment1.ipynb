```python
def get_marks():
    choice = input("Enter marks manually? (y/n): ")

    if choice == "y":
        data = input("Enter marks separated by space: ")
        parts = data.split()
        marks = []

        for item in parts:
            marks.append(int(item))

        return marks
    else:
        return [45, 78, 90, 33, 60]


def analyze_results(marks):
    total = 0
    passed = 0
    failed = 0

    print("\nStudent Results:")

    for i in range(len(marks)):
        mark = marks[i]
        total = total + mark

        if mark >= 50:
            print("Student", i+1, ":", mark, "PASS")
            passed = passed + 1
        else:
            print("Student", i+1, ":", mark, "FAIL")
            failed = failed + 1

    average = total / len(marks)

    print("\nSummary:")
    print("Total Students:", len(marks))
    print("Passed:", passed)
    print("Failed:", failed)
    print("Average:", average)
    print("Highest:", max(marks))
    print("Lowest:", min(marks))
data = get_marks()
analyze_results(data)
```

```python
def get_names():
    choice = input("Enter names manually? (y/n): ").lower()

    if choice == "y":
        raw = input("Enter names (comma separated): ")
        return [n.strip() for n in raw.split(",") if n.strip()]

    return ["Alice","bob","CHARLIE"]

def clean_names(names):

    return [n.strip().lower() for n in names]

def display_results(original, cleaned):
    print("\n--- Results ---")
    print(f"Original: {original}")
    print(f"Cleaned:  {cleaned}")

if __name__ == "__main__":
    names = get_names()
    cleaned = clean_names(names)
    display_results(names, cleaned)
```

```python
def get_messages():
    choice = input("Enter your own messages? y/n: ").strip().lower()

    if choice == 'y':
        raw_input = input("Type messages separated by commas: ")
        messages = [msg.strip() for msg in raw_input.split(',') if msg.strip()]
        return messages
    else:
        return ["Hi", "Welcome to the platform", "OK"]

def analyze_messages(messages):
    print("\n=== Analyzing messages ===")

    long_count = 0
    for msg in messages:
        if len(msg) > 10:
            long_count += 1

    print(f"Found {len(messages)} total messages")
    print(f"{long_count} are longer than 10 chars")
    print()

    for i, msg in enumerate(messages, 1):
        length = len(msg)
        if length > 10:
            print(f"{i}. '{msg}' <- {length} chars (LONG)")
        else:
            print(f"{i}. '{msg}' ({length} chars)")

if __name__ == "__main__":
    msgs = get_messages()
    if not msgs:
        print("No messages? Nothing to analyze.")
    else:
        analyze_messages(msgs)
```

```python
def get_logs():
    choice = input("Want to add your own logs? y/n: ").lower().strip()

    if choice == 'y':
        logs = input("Enter logs (like INFO,ERROR,WARN): ")
        parsed = []
        for log in logs.split(','):
            cleaned = log.strip()
            if cleaned:
                parsed.append(cleaned.upper())
        return parsed

    return ["INFO", "ERROR", "WARNING", "ERROR"]

def analyze_logs(logs):
    if not logs:
        print("No logs dude")
        return
    error_count = 0
    for log in logs:
        if log == "ERROR":
            error_count += 1
    total = len(logs)
    print("\nLog check:")
    print(f"Total logs: {total}")
    print(f"Errors: {error_count}")
    if total > 0:
        print(f"Error %: {error_count/total*100:.0f}%")
    print("\nEach one:")
    for i, log in enumerate(logs, 1):
        print(f"{i}. {log}")
if __name__ == "__main__":
    mylogs = get_logs()
    analyze_logs(mylogs)
```
