```python
class Smart_light:
  def __init__(self,name):
    self.name=name
    self.status=False
  def light_on(self):
    self.status=True
  def light_off(self):
    self.status=False

  def display_status(self):
    if self.status:
      print(f"{self.name} is ON")

    else:
      print(f"{self.name} is OFF")

light=Smart_light("Bedroom Light")
light.light_on()
light.display_status()
```

```python
class Employee:
  def __init__(self,name,id,department):
    self.name=name
    self.id=id
    self.department=department

  def details(self):
    print("Employee ID Card Details:")
    print(f"Employee Name:{self.name}")
    print(f"Employee ID:{self.id}")
    print(f"Employee Department:{self.department}")

detail=Employee("Rahul","EMP102","IT")
detail.details()
```

```python
class Contact:
  def __init__(self,name,phone_no):
    self.name=name
    self.phone_no=phone_no

  def display_status(self):
    print("Contact Saved")
    print(f"Name:{self.name}")
    print(f"Phone:{self.phone_no}")

contacts=Contact("Anita","9876543210")
contacts.display_status()
```

```python
class Store:
  def __init__(self,name,price):
    self.name=name
    self.price=price

  def display_amount(self):
    print(f"Product Name:{self.name}")
    print(f"Price:₹{self.price}")
stores=Store("Headphones","2499")
stores.display_amount()
```

```python
class Movie_rating:
  def __init__(self,name,rating):
    self.name=name
    self.rating=rating

  def display_movie_details(self):
    print(f"Movie Name: {self.name}")
    print(f"Movie Rating: {self.rating}")
movie_detail=Movie_rating("Inception", "4.8/5")
movie_detail.display_movie_details()
```

```python
class Delivery_details:
  def __init__(self,name, address):
    self.name=name
    self.address=address

  def display_delivery_details(self):
    print(f"Delivery Details")
    print(f"Customer Name: {self.name}")
    print(f"Address: {self.address}")

delivery_detail=Delivery_details("Suman", "Hyderabad")
delivery_detail.display_delivery_details()
```
