```python
# Problem 1: Smart Parking Lot Management System

def manage_parking(capacity, logs):
    current_count = 0
    peak_count = 0

    for entry in logs:
        if entry == "IN":
            current_count += 1
        elif entry == "OUT":
            if current_count > 0:
                current_count -= 1

        # Track peak parking usage
        if current_count > peak_count:
            peak_count = current_count

    print("Currently Parked Vehicles:", current_count)

    if current_count > capacity:
        print("Parking Status: Full")
        print("Alert: Capacity Exceeded")
    else:
        print("Parking Status: Available")

    print("Peak Parking Usage:", peak_count)


# Sample Input
capacity = 50
vehicle_logs = ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"]

manage_parking(capacity, vehicle_logs)
```

```python
# Problem 2: Online Food Delivery Time Estimator

def estimate_delivery_time(distance, traffic, weather):
    # Base time calculation (5 minutes per km)
    total_time = distance * 5

    # Add delay based on traffic level
    if traffic == "High":
        total_time += 15
    elif traffic == "Medium":
        total_time += 10
    else:
        total_time += 5

    # Add delay based on weather condition
    if weather == "Rainy":
        total_time += 10
    elif weather == "Storm":
        total_time += 20

    print("Estimated Delivery Time:", total_time, "minutes")


# Sample Input
estimate_delivery_time(8, "High", "Rainy")
```

```python
# Problem 3: Movie Theatre Seat Occupancy Analyzer

def analyze_seats(total_seats, booked_seats):
    booked_count = len(booked_seats)

    # Calculate occupancy percentage
    occupancy_percentage = (booked_count / total_seats) * 100

    print("Occupancy:", int(occupancy_percentage), "%")

    if occupancy_percentage == 100:
        print("Show Status: Housefull")
    elif occupancy_percentage >= 70:
        print("Show Status: Almost Full")
    else:
        print("Show Status: Seats Available")


# Sample Input
total_seats = 200
booked = [1] * 150

analyze_seats(total_seats, booked)
```

```python
# Problem 4: Cloud Server Load Classification System

def classify_server_load(cpu_readings):
    total = 0

    # Calculate total CPU usage
    for value in cpu_readings:
        total += value

    average_load = total / len(cpu_readings)

    print("Average CPU Load:", int(average_load), "%")

    if average_load < 50:
        print("Server Status: Normal")
    elif average_load <= 80:
        print("Server Status: Warning")
    else:
        print("Server Status: Critical")


# Sample Input
cpu_data = [45, 60, 70, 85, 90]
```

```python
# Problem 5: Smart Classroom Resource Usage Monitor

def monitor_resources(resource_usage):
    overused_resources = []

    # Check if any resource is used more than 8 hours
    for resource in resource_usage:
        if resource_usage[resource] > 8:
            overused_resources.append(resource)

    if len(overused_resources) > 0:
        print("Overused Resources:", ", ".join(overused_resources))
        print("Energy Alert: Yes")
    else:
        print("Overused Resources: None")
        print("Energy Alert: No")


# Sample Input
usage_data = {
    "Projector": 6,
    "AC": 9,
    "Lights": 4
}

monitor_resources(usage_data)
```

```python
# Problem 6: Online Event Registration Capacity Controller

def manage_registrations(capacity, total_registrations):

    # Check if registrations exceed capacity
    if total_registrations > capacity:
        confirmed = capacity
        waitlisted = total_registrations - capacity
        status = "Closed"
    else:
        confirmed = total_registrations
        waitlisted = 0
        status = "Open"

    print("Confirmed Registrations:", confirmed)
    print("Waitlisted Users:", waitlisted)
    print("Registration Status:", status)


# Sample Input
manage_registrations(100, 105)
```
