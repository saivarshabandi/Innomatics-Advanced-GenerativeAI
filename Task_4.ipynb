```python
#Task 04: Feb Internship Logic Building Task
#Problem 1: Mobile Recharge Validation System

# Function to validate recharge request
def validate_recharge():

  # List of valid recharge plans
  valid_plans=[199, 299, 399, 599]

  # Loop to allow retry if input is invalid
  while True:
    # Take recharge amount from user
    amount=int(input("Enter recharge amount: "))

    # Check minimum recharge condition
    if amount <= 50:
      print("Recharge amount should be greater than or equal to 50.")

    # Check if amount is one of the valid plans
    elif amount not in valid_plans:
      print("Invalid: Plan not available")

    # If all conditions are satisfied
    else:
      print("Recharge Successful!")
      break

# Call the function
validate_recharge()
```

```python
#Problem 2: Inventory Reorder Alert System
def check_inventory(stock):

    # Loop through each product and its quantity
    for product, quantity in stock.items():
        # If stock is less than 15 â†’ Reorder Alert
        if quantity < 15:
            print(product, "--> Reorder Alert")
        else:
            print(product, "--> Stock OK")

# Dictionary containing product names and their stock quantities
inventory = {
    "Rice": 20,
    "Sugar": 10,
    "Oil": 5,
    "Wheat": 25
}
#call the function
check_inventory(inventory)
```

```python
#Problem 3: Student Result Processing System

def check_result(marks):
  total=0

  # Calculate total marks
  for m in marks:
    total+=m

  # Calculate average
  average=total/len(marks)
  print("Average:", average)

 # Decide the result
  if average >= 50:
    return "Pass"
  else:
    return "Fail"

student_marks = [60, 55, 70, 45]

print("Result:", check_result(student_marks))
```

```python
#Problem 4: Cab Fare Estimator with Retry Option

# Function to calculate cab fare
def cab_calculation(distance, peak):

    # Fixed base fare
    base_fare = 50

    # Rate per kilometer
    rate = 12

    # Calculate normal fare (base fare + distance charge)
    fare = base_fare + (distance * rate)

    # If it is peak hour, add 25% extra
    if peak == "yes":
        fare += fare * 0.25

    # Return final calculated fare
    return fare


# Loop to allow user to retry fare calculation
while True:

    # Take distance input from user
    distance = float(input("Enter distance in km: "))

    # Ask if it is peak hour (convert to lowercase for consistency)
    peak = input("Is it peak hour? (yes/no): ").lower()

    # Call function to calculate fare
    total_fare = cab_calculation(distance, peak)

    # Display total fare
    print("Total Fare: ", total_fare)

    # Ask user if they want to calculate again
    retry = input("Do you want to calculate again?(yes/no): ").lower()

    # Exit loop if user does not want to retry
    if retry != "yes":
        print("Thank you!")
        break
```

```python
#Problem 5: Employee Attendance Eligibility Checker

# Function to check eligibility based on attendance
def check_eligibility(attendance):

    # Variable to count number of present days
    present = 0

    # Loop through attendance list
    for day in attendance:

        # Count only 'P' (Present) days
        if day == "P":
            present += 1

    # Calculate attendance percentage
    percentage = (present / len(attendance)) * 100
    print("Attendance Percentage:", percentage)
    # Check eligibility condition (minimum 75%)
    if percentage >= 75:
        return "Eligible"
    else:
        return "Not Eligible"

attendance_list = ["P", "P", "P", "A", "P"]
print("Status:", check_eligibility(attendance_list))
```

```python
#Problem 6: Password Strength Checker

def check_password(password):
    # Flags to track required conditions
    has_digit = False
    has_special = False

    for ch in password:
        # Check if character is a digit
        if ch.isdigit():
            has_digit = True
       # Check if character is one of the required special characters
        if ch in "@#$":
            has_special = True

    # Final strength check:
    # Minimum length 8 + at least one digit + at least one special character
    if len(password) >= 8 and has_digit and has_special:
        return "Strong Password"
    else:
        return "Weak Password"


password = input("Enter password: ")
print(check_password(password))
```
